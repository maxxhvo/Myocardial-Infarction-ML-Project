```python
feature_cols = model_data_simple.columns.drop(["ZSN", "train_dummy"])

x_train = model_data_simple.loc[model_data_simple["train_dummy"] == 1, feature_cols]
y_train = model_data_simple.loc[model_data_simple["train_dummy"] == 1, "ZSN"]

x_test = model_data_simple.loc[model_data_simple["train_dummy"] == 0, feature_cols]
y_test = model_data_simple.loc[model_data_simple["train_dummy"] == 0, "ZSN"]
```

### Important Hyperparameters to tune:

* max_depth: The maximum depth of the tree
* min_samples_leaf: The minimum number of samples required to be at a leaf node
* max_features: The number of features to consider when looking for the best split

### Additional Stability Parameters:

* n_estimators: The number of trees in the forest
* class_weight: due to class imbalance ~ may set to "balanced"

### Exploratory Grid for Hyperparameter Tuning

```python
# Exploratory Grid for Hyperparameter Tuning

param_grid = {
    "n_estimators": [200, 300, 500],
    "max_depth": [5, 8, 12, None],
    "min_samples_leaf": [1, 5, 10],
    "max_features": ["sqrt", "log2", 0.5],
    "class_weight": [None, "balanced"]
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, average_precision_score

rf = RandomForestClassifier(
    random_state=321,
    n_jobs = 8
)

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=321
)

rf_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_grid,
    n_iter=50,
    scoring="average_precision",   # PR AUC
    cv=cv,
    n_jobs=8,
    random_state=321,
    verbose=1,
    return_train_score=True
)
```

### From the exploratory hyperparameter tuning grid we see:

The following Pattern Repeats across the top 5 performing models:

* max_depth = 5
    * higher depth with other parameters does not seem to improve performance (overfitting)
    * max_depth should be shallow (≈ 4–6)
* min_samples_leaf = 10
    * strong regularization does seem to improve performance (less overfitting)
    * perhaps fit with ~ min_samples_leaf ∈ [5, 10, 20]
* max_features = sqrt
    * reasonable default for our task ~ 0.5 also seems to perform decently

* n_estimators does not seem to have a strong impact on performance; the top performing models have n_estimators = 200, 300, and 500. Fixing n_estimators = 300 is sufficient for our task.

* Class Weight seems to have little impact on performance, likely because the Random Forest is already robust to class imbalance

The Data Overally seems to prefer Smooth and Shallower Trees with Stronger Regularization (lower max_depth, higher min_samples_leaf) to prevent overfitting and improve generalization performance. This is likely due to the relatively small dataset size and high feature dimensionality, which can lead to overfitting if the trees are too deep or not sufficiently regularized.

The cross-validation results indicate that the Random Forest performs best under strong regularization, with shallow trees (max_depth ≈ 3–4) and larger leaf sizes (min_samples_leaf ≈ 12–18) achieving the highest PR AUC (~0.49). Increasing model complexity leads to higher training performance but does not improve validation performance, indicating overfitting. Using max_features = 0.5 further improves generalization, suggesting that additional feature randomness is beneficial. The class_weight parameter does not meaningfully impact performance, as top models consistently use class_weight=None with slightly better PR AUC. This likely reflects the model’s inherent ability to handle moderate class imbalance and the use of PR AUC, which emphasizes ranking over class boundary shifts. Overall, these results suggest that the dataset favors simple, smooth decision boundaries, and that additional model complexity is unlikely to yield meaningful performance gains.

```python
# Predict Y values of trauining data
y_test_pred = best_rf.predict(x_test)
target_names = [' no CHF', 'with CHF']

print('\n------------\n'+
      'Classification report on test data' +
      '\n------------\n')
print(classification_report(y_test, y_test_pred, target_names=target_names))

#Confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_test_pred)
cnf_matrix = pd.DataFrame(cnf_matrix)
cnf_matrix.columns = ['predicted_no_CHF', 'predicted_with_CHF']
cnf_matrix.index = ['true_no_CHF', 'true_with_CHF']

print('\n------------\n'+
      'Confusion matrix on test data' +
      '\n------------\n')

print(cnf_matrix)
```

From the results above, using the default probability threshold of 0.5 is inappropriate in imbalanced clinical prediction tasks because it implicitly assumes equal costs for false positives and false negatives. In this setting, the model achieves high precision but very low recall, meaning many true CHF cases are missed (high false negatives), which is clinically undesirable. Since failing to identify at-risk patients can have serious consequences, the threshold should be lowered to achieve a target recall (e.g., 0.8), ensuring most true cases are captured. This adjustment intentionally increases false positives, but that tradeoff is acceptable in a screening context where additional follow-up is relatively low risk compared to missing a critical condition.

Mathematically, an increased desired recall corresponds to a lower threshold on the predicted probabilities, which allows more positive predictions and thus captures more true positives at the cost of also including more false positives. The optimal threshold can be determined by analyzing the precision-recall curve or using metrics like the F1 score to balance precision and recall according to clinical priorities.

```python
from sklearn.metrics import classification_report, confusion_matrix

y_test_prob = best_rf.predict_proba(x_test)[:, 1]
y_pred_new = (y_test_prob >= chosen_threshold).astype(int)
target_names = [' no CHF', 'with CHF']

print('\n------------\n'+
      'Classification report on test data with adjusted threshold' +
      '\n------------\n')
print(classification_report(y_test, y_pred_new, target_names=target_names))

#Confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred_new)
cnf_matrix = pd.DataFrame(cnf_matrix)
cnf_matrix.columns = ['predicted_no_CHF', 'predicted_with_CHF']
cnf_matrix.index = ['true_no_CHF', 'true_with_CHF']

print('\n------------\n'+
      'Confusion matrix on test data with adjusted threshold' +
      '\n------------\n')

print(cnf_matrix)
```

Lowering the classification threshold increased recall for CHF cases from 0.23 to 0.73, reducing false negatives from 83 to 29. This improvement was accompanied by an increase in false positives and a decrease in precision (0.36), illustrating the inherent tradeoff between sensitivity and specificity. Importantly, selecting the threshold based on training data resulted in a more moderate and generalizable balance compared to the more aggressive threshold derived from test data, while still prioritizing the reduction of missed CHF cases in a clinically relevant manner.

```python
from sklearn.metrics import (
    roc_curve, roc_auc_score,
    precision_recall_curve, average_precision_score
)

dict_eval = {}

for dset in ["train", "test"]:

    if dset == "train":
        x, y = x_train, y_train
    else:
        x, y = x_test, y_test

    # predict probabilities
    y_prob = best_rf.predict_proba(x)[:, 1]

    # ROC curve with thresholds
    fpr, tpr, roc_thresholds = roc_curve(y, y_prob)
    roc_auc = roc_auc_score(y, y_prob)

    # precision-recall curve
    precision, recall, pr_thresholds = precision_recall_curve(y, y_prob)
    pr_auc = average_precision_score(y, y_prob)

    dict_eval[dset] = {
        "fpr": fpr,
        "tpr": tpr,
        "roc_thresholds": roc_thresholds,
        "roc_auc": roc_auc,
        "prec": precision,
        "recall": recall,
        "pr_thresholds": pr_thresholds,
        "pr_auc": pr_auc
    }
```

```python

# Plot ROC curves
plt.plot(dict_eval["train"]["fpr"], dict_eval["train"]["tpr"],
         marker=".", label="Train set")

plt.plot(dict_eval["test"]["fpr"], dict_eval["test"]["tpr"],
         marker=".", label="Test set")

# y=x baseline
plt.plot([0, 1], [0, 1], linestyle="--", color="grey")

anno_list = {}

for dset in ["train", "test"]:

    fpr = dict_eval[dset]["fpr"]
    tpr = dict_eval[dset]["tpr"]
    roc_thresholds = dict_eval[dset]["roc_thresholds"]

    # closest ROC point to chosen threshold
    idx = np.argmin(np.abs(roc_thresholds - chosen_threshold))

    fpr_point = fpr[idx]
    tpr_point = tpr[idx]

    # vertical dotted line at FPR for this threshold
    plt.plot(
        [fpr_point, fpr_point],
        [0, tpr_point],
        linestyle=":",
        color="grey"
    )

    # horizontal dotted line to show recall/TPR
    plt.plot(
        [0, fpr_point],
        [tpr_point, tpr_point],
        linestyle=":",
        color="grey"
    )

    # threshold point
    plt.plot(
        [fpr_point],
        [tpr_point],
        marker="o",
        linestyle="None",
        label=f"{dset} threshold point"
    )

    anno_str = (
    f"{dset}: FPR={fpr_point:.2f}, Recall={tpr_point:.2f}\n"
    f"prob threshold={chosen_threshold:.2f}"
    )

    if dset == "train":
        plt.text(
            fpr_point + 0.02,
            tpr_point + 0.02,
            anno_str,
            fontsize=9,
            bbox=dict(facecolor="white", alpha=0.3, edgecolor="none")

        )
    else:
        plt.text(
            fpr_point - 0.15,
            tpr_point - 0.08,
            anno_str,
            fontsize=9,
            bbox=dict(facecolor="white", alpha=0.3, edgecolor="none")

        )

    anno_list[dset] = f"AUC({dset}): {dict_eval[dset]['roc_auc']:.2f}"



# AUC annotation
auc_str = anno_list["train"] + "\n" + anno_list["test"]
plt.text(0.6, 0.3, auc_str)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC curve for training and test data with adjusted threshold")
plt.legend()
plt.show()
```

```python
# Plot PR curves
plt.plot(
    dict_eval["train"]["recall"],
    dict_eval["train"]["prec"],
    marker=".",
    label="Train set"
)

plt.plot(
    dict_eval["test"]["recall"],
    dict_eval["test"]["prec"],
    marker=".",
    label="Test set"
)

# y=x line.
plt.plot([0,1], [1,0], linestyle='--', color='grey' )

anno_list = {}

for dset in ["train", "test"]:

    recall = dict_eval[dset]["recall"]
    precision = dict_eval[dset]["prec"]
    pr_thresholds = dict_eval[dset]["pr_thresholds"]

    # match chosen threshold to PR curve
    idx = np.argmin(np.abs(pr_thresholds - chosen_threshold))

    recall_point = recall[idx]
    precision_point = precision[idx]

    # vertical line (recall)
    plt.plot(
        [recall_point, recall_point],
        [0, precision_point],
        linestyle=":",
        color="grey"
    )

    # horizontal line (precision)
    plt.plot(
        [0, recall_point],
        [precision_point, precision_point],
        linestyle=":",
        color="grey"
    )

    # threshold point
    plt.plot(
        [recall_point],
        [precision_point],
        marker="o",
        linestyle="None",
        label=f"{dset} threshold point"
    )

    anno_str = (
        f"{dset}: Recall={recall_point:.2f}, Precision={precision_point:.2f}\n"
        f"prob threshold={chosen_threshold:.2f}"
    )

    # offset labels to avoid overlap
    if dset == "train":
        plt.text(
            recall_point - 0.30,
            precision_point + 0.05,
            anno_str,
            fontsize=9,
            bbox=dict(facecolor="white", alpha=0.3, edgecolor="none")
        )
    else:
        plt.text(
            recall_point - 0.35,
            precision_point - 0.12,
            anno_str,
            fontsize=9,
            bbox=dict(facecolor="white", alpha=0.3, edgecolor="none")
        )

    anno_list[dset] = f"PR AUC({dset}): {dict_eval[dset]['pr_auc']:.2f}"

# AUC annotation
auc_str = anno_list["train"] + "\n" + anno_list["test"]
plt.text(
    0.05, 0.15,
    auc_str,
    fontsize=10,
    bbox=dict(facecolor="white", alpha=0.7, edgecolor="none")
)

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall curve for training and test data with adjusted threshold")
plt.legend()
plt.show()
```
